Distribution of comments per video

```python

import csv
from collections import Counter
import matplotlib.pyplot as plt
comment_counts = []
#read csv file
with open('commented_over_0.csv', 'r', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        comment_count = int(row['Comment Count'])
        #append list of comment count
        comment_counts.append(comment_count)
#print maximum comment count       
max_value = max(comment_counts)
print("Maximum value:", max_value)
#sort comment count
comment = sorted(comment_counts, reverse=True)
#plot semilogy
plt.semilogy(comment, 'b-', marker='o')
plt.ylabel("Comments count for per video")
plt.xlabel("rank")
```

Distribution of number of videos of per channel

```python

import csv

video_IDs = []
channel_IDs = []

with open('commented_over_0.csv', 'r', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        video_id = row['Video ID']
        channel_id = row['Channel ID']
        video_IDs.append(video_id)
        channel_IDs.append(channel_id)

# Count the number of videos published by each channel
channel_counts = {}
for channel_id in set(channel_IDs):
    count = channel_IDs.count(channel_id)
    channel_counts[channel_id] = count

# Write the results to a new CSV file
with open('Statistic/video_channel.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Channel ID', 'Video Count'])
    for channel_id, count in channel_counts.items():
        writer.writerow([channel_id, count])

print('Video counts saved to video_counts_per_channel.csv')
        
```

Distribution of user comments number for per video

```python

import csv

video_comments = {}  # Dictionary to store video ID and comment counts for each user

with open('all_comments.csv', 'r', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        video_ID = row['Video ID']
        user_ID = row['User ID']

        # Check if the video ID exists in the dictionary
        if video_ID in video_comments:
            # Check if the user ID exists for the video
            if user_ID in video_comments[video_ID]:
                # Increment the comment count for the user
                video_comments[video_ID][user_ID] += 1
            else:
                # Initialize the comment count for the user
                video_comments[video_ID][user_ID] = 1
        else:
            # Initialize the video ID and comment count for the user
            video_comments[video_ID] = {user_ID: 1}

# Store the comment counts in a separate CSV file
output_file = 'Statistic/comment_counts.csv'
fieldnames = ['Video ID', 'User ID', 'Comment Count']

with open(output_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    
    for video_ID, comments in video_comments.items():
        for user_ID, count in comments.items():
            writer.writerow({'Video ID': video_ID, 'User ID': user_ID, 'Comment Count': count})
```

Duration of videos by year

```python
import matplotlib.pyplot as plt
from datetime import datetime
import csv
#read the duration and publication date
duration_strings = []
publishedDate = []
with open('commented_over_0.csv', 'r', newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        duration_strings.append(row['duration'])
        publishedDate.append(row['Publication Date'])

#convert duration to munites
durations = []
for duration_string in duration_strings:
    try:
        duration = datetime.strptime(duration_string, '%H:%M:%S')
        total_minutes = duration.hour * 60 + duration.minute + duration.second / 60
        durations.append(total_minutes)
    except ValueError:
        print(f"Invalid duration format: {duration_string}")

# Parse the published date strings to datetime objects
published_dates = [datetime.strptime(date, '%Y-%m-%dT%H:%M:%SZ') for date in publishedDate]

# Extract years from published dates
years = [date.year for date in published_dates]

# Create a dictionary to store durations by year
durations_by_year = {}
for year, duration in zip(years, durations):
    if year in durations_by_year:
        durations_by_year[year].append(duration)
    else:
        durations_by_year[year] = [duration]

# Create a list of duration values for each year
duration_values = [durations_by_year[year] for year in sorted(durations_by_year.keys())]

# Create the boxplot
plt.boxplot(duration_values)
plt.yscale('log')
plt.xticks(range(1, len(durations_by_year) + 1), sorted(durations_by_year.keys()), rotation=45) # Rotate the x-axis label by 45 degrees
plt.xlabel('Year')  
plt.ylabel('Minutes')
plt.show()
```
